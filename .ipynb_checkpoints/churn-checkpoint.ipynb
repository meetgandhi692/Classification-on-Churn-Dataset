```python
def uniplot(df,col,title,hue =None):
    
    sns.set_style('whitegrid')
    sns.set_context('talk')
    plt.rcParams["axes.labelsize"] = 20
    plt.rcParams['axes.titlesize'] = 22
    plt.rcParams['axes.titlepad'] = 30
    
    
    temp = pd.Series(data = hue)
    fig, ax = plt.subplots()
    width = len(df[col].unique()) + 7 + 4*len(temp.unique())
    fig.set_size_inches(width , 8)
    plt.xticks(rotation=45)
    plt.yscale('log')
    plt.title(title)
    ax = sns.countplot(data = df, x= col, order=df[col].value_counts().index,hue = hue,palette='bright') 
        
    plt.show()
```

```python
uniplot(new_df1_target1,col='Partner',title='Distribution of Gender for Churned Customers',hue='gender')
```

# Feature engineering

###### Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
y1 = telco['TotalCharges']
print(type(y1))
print(y1.max())
print(y1.min())
y1
```

```python
# (A) Min-Max Scaling (a=1, b=20)
a = 100
b = 200
y1_new = a + ( (y1-min(y1))/(max(y1)-min(y1)) * (b-a) )
print(y1_new.max())
print(y1_new.min())
# df['Fare'] = y1_new
y1_new
```

```python
# (A) Min-Max Scaling (a=0, b=1)
y1_new = (y1-min(y1))/(max(y1)-min(y1))
print(y1_new.max())
print(y1_new.min())
telco['MonthlyCharges'] = y1_new
y1_new
```

```python
print(telco['MonthlyCharges'].max())
print(telco['MonthlyCharges'].min())
```

```python
# (B) Mean Normalization
y1_new = (y1-np.mean(y1))/(max(y1)-min(y1))
print(y1_new.max())
print(y1_new.min())
print(np.mean(y1))
y1_new
```

```python
# (C) Standardization (Z-score Normalization)
y1_new = (y1-np.mean(y1))/np.std(y1)
print(y1_new.max())
print(y1_new.min())
print(np.mean(y1_new))
print(np.std(y1_new))
```

```python
# (D) Robust Scaling
from scipy import stats 
IQR1 = stats.iqr (y1, interpolation = 'midpoint') 
y1_new = (y1-np.median(y1))/IQR1
print(y1_new.max())
print(y1_new.min())
```
